# 5. Perform meaningful visualizations using the dataset. Bring at least 3 
visualizations.

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=iris_data,x='Classification',y='SW').set(title='Relation between Sepal width of different species')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data=iris_data,x='Classification',y='PW').set(title='Petal width according to species ')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(data=iris_data,x='SW',y='PL',hue='Classification')
plt.title("sepal width vs petal length")
plt.show()
```

```python
iris_data.corr()
```

```python
sns.relplot(x=iris_data['PL'],y=iris_data['PW'],col=iris_data['Classification'])
plt.show()
```

```python
plt.bar(iris_data['Classification'],iris_data['SL'])
plt.title('Relation between species and Sepal length')
plt.xlabel("species/classification")
plt.ylabel('Sepal length')
plt.show()
```

```python
sns.pairplot(iris_data,hue='Classification')
plt.show()
```
